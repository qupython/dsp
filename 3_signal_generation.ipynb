# تولید سیگنال سینوسی پیوسته

یک موج سینوسی وقت پیوسته پیوسته نوعی سیگنال است که تناوبی بوده و شکل سینوسی دارد. شکل کلی یک موج سینوسی پیوسته به صورت زیر است:  
$$ x(t) = A \sin(2 \pi f t + \phi) $$  
که در آن A دامنه، f فرکانس، t زمان، و \phi زاویه فاز است. موج سینوسی دوره‌ای به طول  T = 1\f  دارد و هر T ثانیه تکرار می‌شود.  
در پایتون می‌توانید با استفاده از کتابخانه numpy که تابع sin() را برای محاسبه سینوس مقدار داده شده فراهم می‌کند، موج سینوسی CT تولید کنید.  
برای مثال، برای تولید موج سینوسی پیوسته با  A = 1 ،  f = 2، و  \phi = 0  در بازه [0، 1] می‌توانید از کد زیر استفاده کنید:
    <br>
    </span>

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f =2
phi = 0
t = np.linspace(0, 1, 100)
x = A * np.sin(2 * np.pi * f * t + phi)

plt.plot(t, x)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Continuous-Time Sine Wave')
plt.show()
```

این کد یک سیگنال وقت پیوسته (CT) تصادفی تولید می‌کند که شامل نویز سفید با نرخ نمونه‌برداری 1000 هرتز و مدت‌زمان 1 ثانیه است. تابع np.linspace() برای ایجاد یک بردار زمان t استفاده شده است که از 0 شروع شده و تا 1 ثانیه ادامه دارد با گام زمانی 1/fs که معادل فاصله زمانی بین نمونه‌های متوالی است. تابع np.random.normal() برای تولید بردار x از مقادیر تصادفی که از توزیع نرمال با میانگین 0 و انحراف معیار 1 گرفته شده‌اند، به کار رفته است. این مقادیر نمایانگر دامنه سیگنال در هر نقطه زمانی t هستند.برای مثال، برای تولید موج سینوسی پیوسته با  A = 1 ،  f = 2، و  phi = 0  در بازه [0، 1] می‌توانید از کد زیر استفاده کنید:
    <br>
    </span>

```python
import numpy as np
import matplotlib.pyplot as plt
a=2
fs = 1000 # Sampeling
f = 2   # signal freq
phi = 0  # phase
t = np.linspace(0, 1, fs, endpoint=False)

signal =a*np.sin(2*np.pi*f*t+phi)
noise = np.random.normal(0, 1, len(t))
y=signal + 0.1*noise 

plt.plot(t,y)
plt.plot(t,signal)
plt.legend(["noise","main"])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Random CT Signal (White Noise)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000 # sampling frequency
T = 1/fs # sampling period
f = 10 # frequency of sine wave
t = np.arange(0, 1, T) # time vector

x = np.sin(2*np.pi*f*t) # generate sine wave

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine wave')
plt.show()
```

**Square Wave**

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000 # sampling frequency
T = 1/fs # sampling period
f = 5 # frequency of square wave
t = np.arange(0, 1, T) # time vector

x = np.sign(np.sin(2*np.pi*f*t)) # generate square wave

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square wave')
plt.ylim([-1.2, 1.2])
plt.show()
```

Note that the `ylim()` function is used to set the y-axis limits to [-1.2, 1.2] to ensure that the full amplitude range of the square wave is visible in the plot.

**Sawtooth wave**

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

# Define the sampling rate and time vector
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

# Generate the sawtooth wave
f = 10 # Frequency in Hz
x = sig.sawtooth(2*np.pi*f*t)

# Plot the waveform
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Wave ({} Hz)'.format(f))
plt.show()
```

In this code, the `linspace()` function is used to create a time vector t that spans from 0 to 1 second with a step size of 1/fs seconds. The `sawtooth()` function is then used to generate the sawtooth waveform with a frequency of 10 Hz and a period of 1/10 seconds. The resulting waveform is plotted using `matplotlib.pyplot`. The resulting plot shows the sawtooth wave over the time interval [0, 1] seconds.

**Gaussian function**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time vector
t = np.linspace(-5, 5, 1000)

# Define the parameters of the Gaussian function
A = 1 # Amplitude
mu = 0 # Mean
sigma = 1 # Standard deviation

# Generate the Gaussian function
x = A * np.exp(-(t-mu)**2 / (2*sigma**2))

# Plot the waveform
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Gaussian Function')
plt.show()
```

In this code, the `linspace()` function is used to create a time vector t that spans from `-5 to 5` with a step size of `(5-(-5))/1000`. The parameters A, mu, and sigma are used to define the Gaussian function, which is generated using the `np.exp()` function. The resulting waveform is plotted using matplotlib.pyplot. The resulting plot shows the aperiodic Gaussian function over the time interval `[-5, 5]`.

# **Parabolic Signal**
A parabolic signal is a continuous-time signal that has a quadratic form. It can be expressed mathematically as:

`x(t) = At^2 + Bt + C`

where A, B, and C are constants that determine the shape and position of the parabolic signal.

Here's an example Python code to generate a parabolic signal:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
A = 1
B = 2
C = 3

# Generate the time axis
t = np.linspace(-1, 1, 1000)

# Generate the parabolic signal
x = A*t**2 + B*t + C

# Plot the signal
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Parabolic Signal')
plt.grid()
plt.show()
```
